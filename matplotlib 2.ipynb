```python
age=[30,45,26,78,38,25]
engnr=[20000,13000,34000,34567,78934,26378]
dctr=[93787,52672,26785,89365,23456,17835]
tchr=[80000,43567,35678,98703,42567,67355]
bar_width=1
plt.title("Age-Salary Graph")
plt.xlabel("age")
plt.ylabel("Salary")
plt.bar(age,engnr,color='#0000FF',label="Engineer",width = 0.8)
plt.bar(age,dctr,color="000000",label="Doctor",width = 0.8)
plt.bar(age,tchr,color="#EE3B3B",label="Teacher",width = 0.8)
plt.tight_layout()
plt.legend()
plt.show()
```

```python
import numpy as np
```

```python
count=np.arange(len(age))
print(count)
bar_width=0.25


plt.bar(count - bar_width, engnr,width=bar_width,color='#444444',label='Engineer')

plt.bar(count ,dctr,width=bar_width,color='#008df5',label='Doctor')

plt.bar(count +  bar_width,tchr,width=bar_width,color='#e5ae38',label='Teacher')

plt.xticks(ticks=count,labels=age)

plt.title("Age-Salary Graph")
plt.xlabel("age")
plt.ylabel("Salary")

plt.tight_layout()
plt.legend()
plt.show()

```

```python
#stack plot

minutes=[1,2,3,4,5,6,7,8]
player1=[1,3,2,4,3,1,2,1]
player2=[1,3,2,4,6,2,1,2]
player3=[2,4,5,1,2,3,3,1]

labels=['player1','player2','player3']

colors=["#444444","#008df5","#e5ae38"]

plt.stackplot(minutes,player1,player2,player3,labels=labels,colors=colors)
plt.legend(loc='upper left')
plt.title("my stackplot")
plt.xlabel('minutes')
plt.ylabel('players')
plt.tight_layout()
plt.show
```
